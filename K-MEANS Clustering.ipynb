### K-MEANS Clustering:  

K-Means Clustering is an Unsupervised Learning algorithm, which groups the unlabeled dataset into different clusters. Here K defines the number of pre-defined clusters that need to be created in the process, eg. if K=2 then there will be two clusters. 
The main objective of the K-Means algorithm is to minimize the sum of distances between the points and their respective cluster centroid.


Source: https://static.javatpoint.com/tutorial/machine-learning/images/k-means-clustering-algorithm-in-machine-learning.png

```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# create an array of the given shape and 
# populate it with random samples from uniform distribution over [0,1]
X = -2 * np.random.rand(100,2)
X[:10]
```

```python
X1 = 1 + 2 * np.random.rand(50,2)
X1[:10]
```

```python
X[50:100, :] = X1 # copying value of X1 in the 2nd half of X
```

```python
plt.scatter(X[:,0], X[:,1],s = 20,color = 'brown')
plt.show()
```

```python
kmean = KMeans(n_clusters = 4)
kmean.fit(X)
```

```python
kmean.cluster_centers_
```

```python
labels = kmean.labels_
```

```python
plt.scatter(X[:,0],X[:,1],s = 50,c = labels)
plt.scatter(kmean.cluster_centers_[0][0], kmean.cluster_centers_[0][1], s=200, color = 'green')
plt.scatter(kmean.cluster_centers_[1][0], kmean.cluster_centers_[1][1], s=200, color = 'blue')
plt.scatter(kmean.cluster_centers_[2][0], kmean.cluster_centers_[2][1], s=200, color = 'red')
plt.scatter(kmean.cluster_centers_[3][0], kmean.cluster_centers_[3][1], s=200, color = 'BROWN')
plt.show()
```

```python
kmean.predict([[3,1.5]])
```

```python
# Inertia can be recognized as measure of how internally coherent cluster are
kmean.inertia_
```
