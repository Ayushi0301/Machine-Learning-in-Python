### Gradient Descent Method:  
Gradient descent is an iterative optimization algorithm for finding the local minimum of a function. The goal of the gradient descent algorithm is to minimize the given function (say cost function).  
Image Source: Clairvoyant


```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
X  = np.array([2,4,5])
y  = np.array([1.2,2.8,5.3])
```

```python
plt.scatter(X,y)
plt.show()
```

```python
# initialize the parameters
m = 1
c = 0
learning_rate = 0.001
iterations = 1000
error = []
```

```python
for its in range(iterations):# This loop was for iterations
    loss = 0
    cost_m = 0
    cost_c = 0
    for i in range(len(X)): # This loop is for running through all the data points
        y_pred = (m*X[i] + c) #calculating the predicted values
        loss = loss + (y[i] - y_pred)**2 #calulating the loss function
        for j in range(len(X)): # this loop is for calculting the derivative
            partial_dev_wrt_c = -2*(y[j] - (m*X[j] + c)) #partial derivate
            partial_dev_wrt_m = -2*X[j]*(y[j] - (m*X[j] + c)) #partial derivate
            cost_c = cost_c + partial_dev_wrt_c #adding all the dervatives
            cost_m = cost_m + partial_dev_wrt_m #adding all the dervatives

        c = c - cost_c*learning_rate # UPDATING VALUE OF INTERCEPT
        m = m - cost_m*learning_rate # UPDATING VALUE OF INTERCEPT

    error.append(loss)
```

```python
y_pred = m*X+c
```

```python
plt.plot(np.arange(len(error))+1,error,color = 'red')
plt.title('Iterartion vs Error')
plt.xlabel('Iterations')
plt.ylabel('Error')
plt.show()
```
