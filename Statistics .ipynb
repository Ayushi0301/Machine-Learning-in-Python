```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # vizualization library
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mGalarnyk/Python_Tutorials/master/Kaggle/BreastCancerWisconsin/data/data.csv')
```

```python
df
```

```python
df[['diagnosis','area_mean']].tail()
```

```python
df.diagnosis.value_counts()
```

```python
df.boxplot(column = 'area_mean', by = 'diagnosis')
plt.title('')
```

```python
plt.figure(figsize=(8,8))

sns.boxplot(x='diagnosis',y='area_mean',data=df,palette='Set1')
#change the pandas graph to more readable seaborn/matplotlib graph
plt.xlabel('diagnosis',fontsize=45)
plt.ylabel('area_mean',fontsize=45)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
```

```python
mal = df[df.diagnosis=='M']['area_mean']
beng = df[df.diagnosis=='B']['area_mean']
```

```python
plt.figure(figsize=(8,8))
plt.boxplot([mal,beng],notch = True,labels=['M','B'])
plt.show()
```
